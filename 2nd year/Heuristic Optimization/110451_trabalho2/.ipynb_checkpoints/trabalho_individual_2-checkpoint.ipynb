## Dados

```python
import numpy as np

matrix = np.array([[70., 65., 55., 50., 90., 67., 80., 62., 100., 85.],
                   [74., 83., 54., 60., 100., 75., 50., 76., 82., 45.],
                   [71., 87., 66., 58., 74., 81., 48., 52., 100., 64.],
                   [50., 89., 78., 63., 51., 40., 48., 100., 68., 75.],
                   [100., 66., 83., 77., 54., 58., 93., 89., 53., 48.],
                   [80., 55., 70., 65., 94., 47., 60., 88., 73., 60.],
                   [87., 63., 90., 79., 47., 77., 90., 76., 85., 90.],
                   [67., 95., 100., 40., 70., 54., 70., 100., 65., 57.],
                   [90., 45., 88., 48., 65., 68., 80., 46., 71., 67.],
                   [67., 77., 50., 60., 100., 70., 80., 60., 65., 70.]])
```

## 1 iteração

### Lista Tabu de dimensão 2

```python
numero_projects = matrix.shape[1]

# Definir função de avaliação para uma solução
def fitness(solution):
    total_fitness = 0
    for i in range(len(solution)):
        total_fitness += matrix[solution[i]][i]
    return total_fitness

# Definir função para gerar uma solução inicial percorrendo os projetos
def generate_sequential_solution():
    solution = list(range(numero_projects))
    return solution

# Definir a vizinhança da solução
def neighborhood(solution):
    neighborhood = []
    for i in range(numero_projects):
        for j in range(i+1, numero_projects):
            neighbor = solution.copy()
            neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
            neighborhood.append((i, j, neighbor))  
    return neighborhood

def tabu_search(numero_iterations, tabu_size):
    current_solution = generate_sequential_solution()
    best_solution = current_solution.copy()
    best_fitness = fitness(current_solution)
    tabu_list = []
    num_iterations = 0
    
    for _ in range(numero_iterations):
        num_iterations += 1
        neighborhood_list = neighborhood(current_solution)
        best_neighbor = None
        best_neighbor_fitness = float('-inf')
        best_neighbor_indices = None
        
        # Ordenar a lista de vizinhança por ordem decrescente de aptidão
        neighborhood_list = sorted(neighborhood_list, key=lambda x: fitness(x[2]), reverse=True)
        
        for i, j, neighbor in neighborhood_list:
            if (i, j) not in tabu_list:
                best_neighbor = neighbor
                best_neighbor_fitness = fitness(neighbor)
                best_neighbor_indices = (i, j)
                break  
        
        current_solution = best_neighbor
        current_fitness = best_neighbor_fitness
        
        if best_neighbor_indices not in tabu_list:
            if len(tabu_list) < tabu_size:
                tabu_list.append(best_neighbor_indices)
            else:
                tabu_list.pop(0)
                tabu_list.append(best_neighbor_indices)
        
        if current_fitness > best_fitness:
            best_solution = current_solution
            best_fitness = current_fitness
            
    print("Projetos atribuídos aos cientistas: \n")
    for i, projeto in enumerate(best_solution):
        cientista = "C" + str(projeto + 1)
        print(f"Projeto P{i + 1}: {cientista}")
    print("\n")
            
    return best_solution, best_fitness, num_iterations, tabu_list

best_solution, best_fitness, num_iterations, tabu_list = tabu_search(1, 2)

best_solution_print = [projeto + 1 for projeto in best_solution]

# Incrementar em 1 cada componente do par na lista tabu

tabu_list_print = [(i+1, j+1) for i, j in tabu_list]


print("Melhor solução: ", best_solution_print)
print("Aptidão total: ", best_fitness.round(0).astype(int))
print("Número de iterações: ", num_iterations)
print("\nLista Tabu:", tabu_list_print)
```

## 100 iterações

#### Código

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

numero_projects = matrix.shape[1]

# Definir função de avaliação para uma solução
def fitness(solution):
    total_fitness = 0
    for i in range(len(solution)):
        total_fitness += matrix[solution[i]][i]
    return total_fitness

# Definir função para gerar uma solução inicial percorrendo os projetos
def generate_sequential_solution():
    solution = list(range(numero_projects))
    return solution

# Definir a vizinhança da solução
def neighborhood(solution):
    neighborhood = []
    for i in range(numero_projects):
        for j in range(i+1, numero_projects):
            neighbor = solution.copy()
            neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
            neighborhood.append((i, j, neighbor))  
    return neighborhood

def tabu_search(numero_iterations, tabu_size):
    current_solution = generate_sequential_solution()
    best_solution = current_solution.copy()
    best_fitness = fitness(current_solution)
    tabu_list = []
    num_iterations = 0
    fitness_history = []  # Lista para armazenar a aptidão total em cada iteração
    tabu_usage = {}  # Dicionário para armazenar a frequência de uso de movimentos tabu
    
    iterations_data = []  # Lista para armazenar dados das iterações
    
    for _ in range(numero_iterations):
        num_iterations += 1
        neighborhood_list = neighborhood(current_solution)
        best_neighbor = None
        best_neighbor_fitness = float('-inf')
        best_neighbor_indices = None
        
        neighborhood_list = sorted(neighborhood_list, key=lambda x: fitness(x[2]), reverse=True)
        
        for i, j, neighbor in neighborhood_list:
            if (i, j) not in tabu_list:
                best_neighbor = neighbor
                best_neighbor_fitness = fitness(neighbor)
                best_neighbor_indices = (i, j)
                break
        
        current_solution = best_neighbor
        current_fitness = best_neighbor_fitness
        
        if best_neighbor_indices not in tabu_list:
            if len(tabu_list) < tabu_size:
                tabu_list.append(best_neighbor_indices)
            else:
                tabu_list.pop(0)
                tabu_list.append(best_neighbor_indices)
        
        if current_fitness > best_fitness:
            best_solution = current_solution
            best_fitness = current_fitness
        
        fitness_history.append(current_fitness)
        tabu_usage[best_neighbor_indices] = tabu_usage.get(best_neighbor_indices, 0) + 1
        
        # Armazenar dados da iteração
        iteration_data = {
            'Iteração': num_iterations,
            'Solução': [projeto + 1 for projeto in best_solution.copy()],
            'Aptidão Total': round(current_fitness, 0).astype(int),
            'Lista Tabu': [(i+1, j+1) for i, j in tabu_list.copy()]
        }
        iterations_data.append(iteration_data)
        
    df_iterations = pd.DataFrame(iterations_data)
    
    plt.plot(range(1, num_iterations + 1), fitness_history)
    plt.xlabel('Iteração')
    plt.ylabel('Aptidão Total')
    plt.title('Evolução da Aptidão Total')
    plt.show()

    moves = list(tabu_usage.keys())
    frequencies = list(tabu_usage.values())

    moves = [(x + 1, y + 1) for x, y in moves]
    fig, ax = plt.subplots(figsize=(10, 6))
    xtick_labels = [f"({move[0]}, {move[1]})" for move in moves]
    plt.bar(range(len(moves)), frequencies)
    plt.xlabel('Movimentos Tabu')
    plt.ylabel('Frequência de Uso')
    plt.title('Frequência de Uso de Movimentos Tabu')
    plt.xticks(range(len(moves)), xtick_labels, rotation=45, ha='right')
    plt.tick_params(axis='x', which='major', pad=10)
    plt.show()

    print("Projetos atribuídos aos cientistas: \n")
    for i, projeto in enumerate(best_solution):
        cientista = "C" + str(projeto + 1)
        print(f"Projeto P{i + 1}: {cientista}")
    print("\n")
            
    return best_solution, best_fitness, num_iterations, tabu_list, df_iterations
```

### Lista Tabu de dimensão 2

```python
best_solution, best_fitness, num_iterations, tabu_list, df_iterations = tabu_search(100, 2)

df_iterations = df_iterations.set_index('Iteração')
df_iterations.index.name = 'Iteração'

tabu_list_print = [(i+1, j+1) for i, j in tabu_list]
best_solution_print = [projeto + 1 for projeto in best_solution]

print("Melhor solução: ", best_solution_print)
print("Aptidão total: ", best_fitness.round(0).astype(int))
print("Número de iterações: ", num_iterations)
print("\nLista Tabu:", tabu_list_print)

# Imprimir dataframe das iterações
print("\nDataframe das iterações:")
df_iterations
```
