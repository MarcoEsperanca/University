### d) Crossover por um Ponto

```python
import numpy as np
def crossover(pai1, pai2, m = 4):
    
    # a divisão (m) não pode ser igual ao superior ao tamanho do cromossoma, quer seja pai1 ou pai2
    if m >= len(pai1) and m >= len(pai2):
        print(len(pai1))
        print(len(pai2))
        raise TypeError("Inválido: Valor de m inválido!")
    
    # o mesmo se aplica para o tamanho dos cromossomas, que tem de ser igual
    if len(pai1) != len(pai2):
        raise TypeError("Inválido: cromossomas com tamanhos diferentes!")
    
    # criação base para os cromossomas filhos, inicialmente sem valores atribuidos
    filho1 = np.zeros(len(pai1), dtype = int)
    filho2 = np.zeros(len(pai1), dtype = int)
    
    
    for i in range(len(filho1)):
        if i < m:
            filho1[i] = pai1[i]
        else:
            filho1[i] = pai2[i]
    
    for k in range(len(filho2)):
        if k < m:
            filho2[k] = pai2[k]
        else:
            filho2[k] = pai1[k]

    return filho1, filho2
```

### Geração da população

```python
import random
import numpy as np

# Definir a semente
random.seed(777) 

# Definir os dados das tarefas e tempos de processamento
tarefas = {
    'T1': 38,
    'T2': 33,
    'T3': 36,
    'T4': 20,
    'T5': 32,
    'T6': 29,
    'T7': 46,
    'T8': 38,
    'T9': 34,
    'T10': 40
}

# Definir a dimensão da população
dimensao_populacao = 200

# Definir uma função para calcular o tempo total de processamento em cada máquina
def calcular_tempo_total(cromossoma):
    m1 = sum(tarefas[tarefa] for tarefa in cromossoma[:4])
    m2 = sum(tarefas[tarefa] for tarefa in cromossoma[4:7])
    m3 = sum(tarefas[tarefa] for tarefa in cromossoma[7:])

    tempo_total = [m1, m2, m3]
    return tempo_total

# Definir uma função para gerar um cromossoma aleatório, considerando as restrições para a conclusão de T3, T4 e T5
def gerar_cromossoma():
    cromossoma = random.sample(list(tarefas.keys()), len(tarefas))
    
    idx = None
    for i, tarefa in enumerate(cromossoma[:4]):
        if tarefa in ['T3', 'T4', 'T5']:
            idx = i
            break

    if idx is not None and sum([tarefas[tarefa] for tarefa in cromossoma[:idx+1]]) > 96:
        return gerar_cromossoma()  # Chamada recursiva para gerar um novo cromossoma, se violar a condição
    
    idx2 = None
    for i, tarefa in enumerate(cromossoma[4:7], start=4):
        if tarefa in ['T3', 'T4', 'T5']:
            idx2 = i
            break

    if idx2 is not None and sum([tarefas[tarefa] for tarefa in cromossoma[4:idx2+1]]) > 96:
        return gerar_cromossoma()  # Chamada recursiva para gerar um novo cromossoma,  se violar a condição
    
    idx3 = None
    for i, tarefa in enumerate(cromossoma[7:], start=7):
        if tarefa in ['T3', 'T4', 'T5']:
            idx3 = i
            break

    if idx3 is not None and sum([tarefas[tarefa] for tarefa in cromossoma[7:idx3+1]]) > 96:
        return gerar_cromossoma()  # Chamada recursiva para gerar um novo cromossoma, se violar a condição
    
    if idx is not None and idx2 is not None and idx3 is not None:
        return cromossoma
    
    return gerar_cromossoma()  # Chamada recursiva se alguma posição não foi preenchida corretamente

# Criar a população inicial e calcular o tempo de cada cromossoma para cada máquina
populacao_inicial = []
for _ in range(dimensao_populacao):
    cromossoma = gerar_cromossoma()
    tempo_total = calcular_tempo_total(cromossoma)
    populacao_inicial.append((cromossoma, tempo_total))

# Exibir a população inicial
for i, (cromossoma, tempo_total) in enumerate(populacao_inicial):
    print(f"Cromossoma {i+1}: {cromossoma} | Tempo Total por máquina: {tempo_total} | \nTempo máximo gasto numa máquina: {np.max(tempo_total)}")
```

### Selecionação dos Pais por Torneio

```python
# Definir a semente
random.seed(777) 

# k representa o número de cromossomas selecionados da população de forma aleatória
k = 200

# Função para a seleção por torneio
def selecao_por_torneio(populacao, k):
    participantes = random.sample(populacao, k)
    melhor_cromossomo = min(participantes, key=lambda c: c[1])
    return melhor_cromossomo[0]

# Criar a população de exemplo
populacao = []
for _ in range(dimensao_populacao):
    cromossomo = gerar_cromossoma()
    tempo_total = calcular_tempo_total(cromossomo)
    populacao.append((cromossomo, tempo_total))

# Seleção do Pai1 por torneio
pai1 = selecao_por_torneio(populacao, k)

# Seleção do Pai2 por torneio, garantindo que é diferente do Pai1, se não não faz sentido
pai2 = selecao_por_torneio(populacao, k)
while pai2 == pai1:
    pai2 = selecao_por_torneio(populacao, k)
```

### Crossover a um ponto e a garantia de admissibilidade

```python
import random
import numpy as np

# m = 4
def crossover(pai1, pai2, m=4):
    if m >= len(pai1) and m >= len(pai2):
        return "Inválido: Escolha um valor de m, considerando a dimensão do cromossoma!"
    
    if len(pai1) != len(pai2):
        raise ValueError("Inválido: os cromossomas têm que ter a mesma dimensão!")
    
    filho1 = np.zeros(len(pai1), dtype='<U3')
    filho2 = np.zeros(len(pai1), dtype='<U3')
    
    for i in range(len(filho1)):
        if i < m:
            filho1[i] = pai1[i]
        else:
            filho1[i] = pai2[i]
    
    for k in range(len(filho2)):
        if k < m:
            filho2[k] = pai2[k]
        else:
            filho2[k] = pai1[k]
    
    # Reparo
    # tarefas repetidas
    tarefas_repetidas_filho1 = [tarefa for tarefa in filho1 if np.count_nonzero(filho1 == tarefa) > 1]
    tarefas_repetidas_filho2 = [tarefa for tarefa in filho2 if np.count_nonzero(filho2 == tarefa) > 1]
    
    for tarefa in tarefas_repetidas_filho1:
        indices_repetidos = np.where(filho1 == tarefa)[0]
        for indice in indices_repetidos[1:]:
            tarefas_disponiveis = np.setdiff1d(list(tarefas.keys()), filho1)
            filho1[indice] = random.choice(tarefas_disponiveis)
    
    for tarefa in tarefas_repetidas_filho2:
        indices_repetidos = np.where(filho2 == tarefa)[0]
        for indice in indices_repetidos[1:]:
            tarefas_disponiveis = np.setdiff1d(list(tarefas.keys()), filho2)
            filho2[indice] = random.choice(tarefas_disponiveis)
    
    # Verificar se T3, T4 e T5 são entregues a tempo
    if sum([tarefas[tarefa] for tarefa in filho1]) > 96:
        idx = None
        for i, tarefa in enumerate(filho1):
            if tarefa in ['T3', 'T4', 'T5']:
                idx = i
                break
        if idx is not None:
            tarefas_disponiveis = np.setdiff1d(list(tarefas.keys()), filho1)
            if len(tarefas_disponiveis) > 0:
                filho1[idx] = random.choice(tarefas_disponiveis)
    
    if sum([tarefas[tarefa] for tarefa in filho2]) > 96:
        idx = None
        for i, tarefa in enumerate(filho2):
            if tarefa in ['T3', 'T4', 'T5']:
                idx = i
                break
        if idx is not None:
            tarefas_disponiveis = np.setdiff1d(list(tarefas.keys()), filho2)
            if len(tarefas_disponiveis) > 0:
                filho2[idx] = random.choice(tarefas_disponiveis)
    
    return filho1, filho2
```

### Mutação por Troca e garantia de admissibilidade

```python
import random
import numpy as np

# semente
random.seed(777)

indices_mutacao = []  # Lista para armazenar os índices dos cromossomos mutados
probabilidade_mutacao = 0.1  # Probabilidade de mutação (pm)

# Iterar sobre a população inicial
for i, (cromossoma, tempo_total) in enumerate(populacao_inicial):
    # Verificar se a mutação deve ocorrer com base na probabilidade
    if random.random() < probabilidade_mutacao:
        # Selecionar dois índices aleatórios
        indice1, indice2 = random.sample(range(len(cromossoma)), 2)

        # Trocar os valores dos genes nos índices selecionados
        cromossoma[indice1], cromossoma[indice2] = cromossoma[indice2], cromossoma[indice1]

        # Verificar se as tarefas T3, T4 e T5 estão presentes no cromossoma
        t3_presente = 'T3' in cromossoma
        t4_presente = 'T4' in cromossoma
        t5_presente = 'T5' in cromossoma

        # Calcular o novo tempo total após a mutação
        tempo_total = calcular_tempo_total(cromossoma)

        # Verificar se as tarefas T3, T4 e T5 são executadas a tempo - garantir a admissibilidade
        if sum([tarefas[tarefa] for tarefa in cromossoma if tarefa in ['T3', 'T4', 'T5']]) > 96:
            idx = None
            for j, tarefa in enumerate(cromossoma):
                if tarefa in ['T3', 'T4', 'T5']:
                    idx = j
                    break
            if idx is not None:
                tarefas_disponiveis = np.setdiff1d(list(tarefas.keys()), cromossoma)
                if len(tarefas_disponiveis) > 0:
                    cromossoma[idx] = random.choice(tarefas_disponiveis)
                    tempo_total = calcular_tempo_total(cromossoma)
                    if tempo_total > 96:
                        # Reverter a mutação
                        cromossoma[indice1], cromossoma[indice2] = cromossoma[indice2], cromossoma[indice1]
                else:
                    # Selecionar outra posição aleatória para trocar
                    outra_posicao = random.choice(range(len(cromossoma)))
                    cromossoma[idx], cromossoma[outra_posicao] = cromossoma[outra_posicao], cromossoma[idx]
                    tempo_total = calcular_tempo_total(cromossoma)
                    if tempo_total > 96:
                        # Reverter a mutação
                        cromossoma[indice1], cromossoma[indice2] = cromossoma[indice2], cromossoma[indice1]

        # Adicionar o índice do cromossoma mutado à lista indices_mutacao
        indices_mutacao.append(i)
```
